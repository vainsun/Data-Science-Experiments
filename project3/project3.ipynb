This suggests that the majority of companies have a low level of digital transformation, while those with high digital transformation are relatively rare.

To calculate the mean of DDT values grouped by year in a dataset

```python

mean_ddt_by_year = df.groupby('year')['DDT'].mean()
print(mean_ddt_by_year)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(mean_ddt_by_year.index, mean_ddt_by_year.values, marker='o')
plt.title('the mean of DDT values grouped by year in a dataset')
plt.xlabel('year')
plt.ylabel('mean of DDT')
plt.grid(True)
plt.show()
```

This graph shows the trend of the mean DDT values from 2013 to 2023. Overall, the mean DDT values have increased over the years. Starting at around 9 in 2013, they rose steadily to exceed 30 in 2021. Although there was a slight decline in 2022, the values rebounded in 2023 and remained high.
